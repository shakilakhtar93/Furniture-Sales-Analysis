# Percentage lost analysis

```python
a2 = north_100['order phase']=='Lost'
a3 = north_100.loc[a2]

b2 = east_100['order phase']=='Lost'
b3 = east_100.loc[b2]

c2 = west_100['order phase']=='Lost'
c3 = west_100.loc[c2]

d2 = ncr_100['order phase']=='Lost'
d3 = ncr_100.loc[d2]

e2 = bengaluru_100['order phase']=='Lost'
e3 = bengaluru_100.loc[e2]

f2 = hyderabad_100['order phase']=='Lost'
f3 = hyderabad_100.loc[f2]

g2 = chennai_100['order phase']=='Lost'
g3 = chennai_100.loc[g2]

h2 = east_central_100['order phase']=='Lost'
h3 = east_central_100.loc[h2]

i2 = west_central_100['order phase']=='Lost'
i3 = west_central_100.loc[i2]
```

```python
all1 = [a3['sales'].sum(),b3['sales'].sum(),c3['sales'].sum(),d3['sales'].sum(),e3['sales'].sum(),f3['sales'].sum(),g3['sales'].sum(), h3['sales'].sum(), i3['sales'].sum()]
index = ['north', 'east', 'west', 'ncr','bengaluru', 'hyderabad', 'chennai', 'east_central', 'west_central']
sales_100 = pd.DataFrame(data=all1, index=index, columns=['Lost Sales'])
sales_100
```

```python
s1 = north_100['sales'].sum()
s2 = east_100['sales'].sum()
s3 = west_100['sales'].sum()
s4 = ncr_100['sales'].sum()
s5 = bengaluru_100['sales'].sum()
s6 = hyderabad_100['sales'].sum()
s7 = chennai_100['sales'].sum()
s8 = east_central_100['sales'].sum()
s9 = west_central_100['sales'].sum()
total_all=[s1, s2, s3, s4, s5, s6, s7, s8, s9]
```

```python
sales_100['Total_sales'] = total_all
sales_100
```

```python
sales_100['Final_sales']=(sales_100['Total_sales']-sales_100['Lost Sales'])
sales_100
```

```python
sales_100['% loss']=(sales_100['Total_sales']-sales_100['Final_sales'])/sales_100['Total_sales']*100
sales_100
```

```python
plt.figure(figsize=(8,8))
plt.pie(sales_100['% loss'], labels=sales_100.index, autopct='%.2f%%', pctdistance=0.5)
plt.legend(loc='upper right')
plt.show()
```

```python
plt.style.use('fivethirtyeight')
sales_100['% loss'].sort_values(ascending=False).plot(kind='bar', color='blue')
sales_100['% loss'].sort_values(ascending=False).plot(kind='line', color='green')
plt.xticks(rotation=45)
plt.show()
```
